```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2 * np.pi, 2 * np.pi, 200, endpoint=True)
y = np.cos(x)

plt.plot(x, y)
plt.title("Cosine Function")
plt.xlabel("x")
plt.ylabel("cos(x)")
plt.grid(True)
plt.show()
```
