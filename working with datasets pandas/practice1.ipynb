```python
# Import the data into Python environment as a Pandas DataFrame.

import pandas as pd
import numpy as np

df = pd.read_csv(r'C:\Users\patha\Downloads\rainfall1627650742214.csv')
df
```

```python
# Check for missing values, if any and drop the corresponding rows

df.dropna(inplace = True)
df.info()
```

```python
# Find the district that gets the highest annual rainfall

#Using list comprehension to get the numerical columns
list1 = [col for col in df.columns if df[col].dtype in ['float', 'int64']]
df[list1].agg(['max'])
```

```python
# Display the top 5 states that get the highest annual rainfall.

df['ANNUAL'].nlargest(n=5)
```

```python
# Drop the columns 'Jan-Feb', 'Mar-May', 'Jun-Sep', 'Oct-Dec'

df= df.drop(['Jan-Feb','Mar-May', 'Jun-Sep', 'Oct-Dec'],axis=1)
print(df)
```

```python
# Display the state-wise mean rainfall for all the months using a pivot table.

import pandas as pd

# Grouping by state and district and calculating mean of ANNUAL
grouped_mean = df.groupby(['STATE_UT_NAME', 'DISTRICT'])['ANNUAL'].mean()

# Convert the result to a DataFrame and rename the column
grouped_mean = grouped_mean.reset_index()
grouped_mean.columns = ['STATE_UT_NAME', 'DISTRICT', 'annual_mean']

# View result
grouped_mean.head()

```

```python
pivot1 = pd.pivot_table(df, index = 'ANNUAL', aggfunc=np.mean)
pivot1
```

```python
# Display the count of districts in each state.

district_count = df.groupby('STATE_UT_NAME')['DISTRICT'].nunique().reset_index()
district_count.columns = ['STATE_UT_NAME', 'district_count']
print(district_count)
```

```python
# For each state, display the district that gets the highest rainfall in May. 
# Also display the recorded rainfall.

# Find the district with the maximum May rainfall in each state
idx = df.groupby('STATE_UT_NAME')['MAY'].idxmax()

# Use .loc to get those rows
max_rainfall_may = df.loc[idx, ['STATE_UT_NAME', 'DISTRICT', 'MAY']].reset_index(drop=True)

# Rename column for clarity
max_rainfall_may.columns = ['STATE_UT_NAME', 'DISTRICT', 'max_may_rainfall']

# Display result
print(max_rainfall_may)
```
